```python
# Standard 
import numpy as np 
import pandas as pd
from numpy.random import randn 

#stats
from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')

tips
```

```python
sns.lmplot('total_bill',
           'tip',tips)
```

```python
sns.lmplot('total_bill','tip',tips,order=4,
           scatter_kws={'marker':'o','color':'indianred'},line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot('total_bill','tip',tips,fit_reg=False)
```

```python
tips.head()
```

```python
tips['tip_perc'] = 100*(tips['tip']/tips['total_bill']) 
```

```python
tips.head()
```

```python
sns.lmplot(data=tips,x='size',y='tip_perc')
```

```python
sns.lmplot(data=tips,x='size',y='tip_perc',x_jitter =.1)
```

```python
sns.lmplot(data=tips,x='size',y='tip_perc',x_estimator=np.mean)
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip_perc',hue='sex',markers=['x','o'])
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip_perc',hue='day')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip_perc',lowess=True,line_kws={'color':'black'})
```

```python
sns.regplot(data=tips,x='total_bill',y='tip_perc')
```

```python
fig, (axis1,axis2) = plt.subplots(1,2,sharey=True)

sns.regplot(data=tips,x='total_bill',y='tip_perc',ax=axis1)
sns.violinplot(data=tips,y='tip_perc',x='size',color='red',ax=axis2)
```
