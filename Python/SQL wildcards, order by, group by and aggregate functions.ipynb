```python
# Imports
import sqlite3
import pandas as pd
con = sqlite3.connect("sakila.db")

# Set function as our sql_to_pandas

def sql_to_df(sql_query):

    # Use pandas to pass sql query using connection form SQLite3
    df = pd.read_sql(sql_query, con)

    # Show the resulting DataFrame
    return df
```

```python
query = '''SELECT COUNT(customer_id) FROM customer;'''
sql_to_df(query).head()
```

```python
query = '''SELECT * FROM customer WHERE first_name LIKE 'M%'; '''
sql_to_df(query).head()
```

```python
query = '''SELECT * FROM customer WHERE last_name LIKE '_ING'; '''
sql_to_df(query).head()
```

```python
query = '''SELECT * FROM customer WHERE first_name GLOB '[AB]*'; '''
sql_to_df(query).head()
```

```python
query = '''SELECT * FROM customer WHERE first_name LIKE 'M%' ORDER BY customer_id DESC; '''
sql_to_df(query).head()
```

```python
query = '''SELECT * FROM customer WHERE first_name LIKE 'M%' GROUP BY address_id ; '''
sql_to_df(query).head()
```
