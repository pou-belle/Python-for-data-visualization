```python
# Standard 
import numpy as np 
import pandas as pd
from numpy.random import randn 

#stats
from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
dataset=randn(25)
sns.rugplot(dataset)
plt.ylim(0,1)
```

```python
plt.hist(dataset,alpha=0.3)
sns.rugplot(dataset)
```

```python
sns.rugplot(dataset)
x_min = dataset.min()-2
x_max =dataset.min()+2
x_axis =np.linspace(x_min,x_max,100)
bandwith = ((4*dataset.std()**5)/(3*len(dataset)))**0.2
kernel_list =[]
for datapoint in dataset:
    
    #create kernel for each poitnt and append to the list 
    kernel = stats.norm(datapoint,bandwith).pdf(x_axis)
    kernel_list.append(kernel)
    #scale for plotting 
    kernel= kernel/kernel.max()
    kernel =kernel*0.4
    
    plt.plot(x_axis,kernel,color='grey',alpha = 0.5)
    
    
    
    
```

```python
 sum_kde = np.sum(kernel_list, axis=0)
fig=plt.plot(x_axis,sum_kde,color='indianred')
sns.rugplot(dataset)
plt.yticks([])
plt.suptitle("sum of the basis function")
```

```python
sns.kdeplot(dataset)
bandwith = ((4*dataset.std()**5)/(3*len(dataset)))**0.2
```

```python
sns.rugplot(dataset,color='black')

for bw in np.arange(0.5,2, 0.5):
    sns.kdeplot(dataset,bw=bw,lw=1.8, label=bw)
```

```python
kernel_options= ['bw','cos','epa','gaus','tri','triw']

for kern in kernel_options:
    sns.kdeplot(dataset,kernel=kern,label=kern,bw_method='scott')
```

```python
sns.kdeplot(dataset,cumulative=True)
```
